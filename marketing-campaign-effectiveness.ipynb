```python
import pandas as pd
df=pd.read_csv("Digital_marketing_dataset.csv")
df.head()
```

```python
print(df["CampaignChannel"].value_counts())
print(df["CampaignType"].value_counts())
print(df["AdvertisingPlatform"].value_counts())
df.describe()
```

```python
import matplotlib.pyplot as plt
```

```python
channel_metrics=df.groupby("CampaignChannel")[['ConversionRate', 'ClickThroughRate', 'AdSpend']].mean().sort_values(by='ConversionRate', ascending=False)
print(channel_metrics)
```

```python
plt.figure(figsize=(8,5))
channel_metrics['ConversionRate'].plot(kind='bar', color='skyblue')
plt.title('Average Conversion Rate by Campaign Channel')
plt.ylabel('Average Conversion Rate')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
channel_metrics['ClickThroughRate'].plot(kind='bar', color='orange')
plt.title('Average Click Through Rate by Campaign Channel')
plt.ylabel('Average CTR')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
channel_metrics['AdSpend'].plot(kind='bar', color='green')
plt.title('Average Ad Spend by Campaign Channel')
plt.ylabel('Average Ad Spend')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
```

```python
df["AgeGroup"]=pd.cut(df["Age"],bins=[18,25,35,45,60,100], labels=["18-28", "26-35","36-45","46-60","60+"])
age_group_conv=df.groupby("AgeGroup")["ConversionRate"].mean()
age_group_conv.plot(kind="bar", color="violet")
plt.title("Average conversion Rate by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Conversion Rate")
plt.show()
```

```python
gender_conv = df.groupby('Gender')['ConversionRate'].mean()
gender_conv.plot(kind='bar', color='dodgerblue')
plt.title('Average Conversion Rate by Gender')
plt.ylabel('Conversion Rate')
plt.show()
```

```python
df['IncomeGroup'] = pd.qcut(df['Income'], 5, labels=['Low','BelowAvg','Average','AboveAvg','High'])
income_conv = df.groupby('IncomeGroup')['ConversionRate'].mean()
income_conv.plot(kind='bar', color='seagreen')
plt.title('Average Conversion Rate by Income')
plt.xlabel('Income Group')
plt.ylabel('Conversion Rate')
plt.show()
```

```python
df['EstimatedRevenue'] = df['Conversion'] * 100
df['ROI'] = (df['EstimatedRevenue'] - df['AdSpend']) / df['AdSpend']
roi_by_channel = df.groupby('CampaignChannel')['ROI'].mean().sort_values(ascending=False)
print(roi_by_channel)
```

```python
plt.figure(figsize=(8,5))
roi_by_channel.plot(kind='bar', color='navy')
plt.title('Average ROI by Campaign Channel')
plt.ylabel('Average ROI')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
roi_by_platform = df.groupby('AdvertisingPlatform')['ROI'].mean().sort_values(ascending=False)
roi_by_platform.plot(kind='bar', color='purple')
plt.title('Average ROI by Advertising Platform')
plt.ylabel('Average ROI')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
roi_by_channel.plot(kind='bar', color='navy')
plt.title('ROI by Campaign Channel')
plt.ylabel('ROI')
```

```python
plt.subplot(1,3,2)
channel_metrics['ConversionRate'].plot(kind='bar', color='blue')
plt.title('Conversion Rate by Channel')
plt.ylabel('Conversion Rate')
```

```python
plt.subplot(1,3,3)
channel_metrics['ClickThroughRate'].plot(kind='bar', color='teal')
plt.title('CTR by Channel')
plt.ylabel('Click Through Rate')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
age_group_conv.plot(kind='bar', color='violet')
plt.title('Conversion Rate by Age Group')

plt.subplot(1,3,2)
gender_conv.plot(kind='bar', color='dodgerblue')
plt.title('Conversion Rate by Gender')

plt.subplot(1,3,3)
income_conv.plot(kind='bar', color='seagreen')
plt.title('Conversion Rate by Income')
plt.tight_layout()
plt.show()
```


Marketing Campaign Effectiveness Project   Summary

This project analyzes a comprehensive digital marketing campaign dataset, focusing on understanding which strategies deliver the highest conversion and ROI. Using Python (pandas, matplotlib), insights were extracted to guide future campaign investments and targeting.

Key Findings

Highest ROI Channels: Certain campaign channels consistently outperformed others, delivering better conversion rates and return on investment for the same ad spend.

Channel Performance: Social media and search platforms often see greater click-through and conversion rates, while email campaigns are more cost-effective for some customer segments.

Demographic Insights:

Age: Middle-aged groups (26–45) have higher conversion rates compared to younger or older cohorts.

Gender: Subtle differences in average conversion rate highlight opportunities for tailored messaging.

Income: Customers in 'Above Average' and 'High' income brackets are more likely to convert.

Platform ROI: Advertising platforms show distinct cost-effectiveness; for example, Google and Facebook provide higher average ROI compared to traditional channels.

Campaign Type: Discount and event campaigns show spikes in engagement but may lead to lower margins. Brand-building campaigns deliver longer-term ROI.

Business Recommendations

Allocate more ad spend to channels and platforms with historically high ROI.

Target higher-income and prime-age demographics for premium sales or special offers.

Increase A/B testing of campaign types to find the best balance between volume and margin.

Use dashboards to monitor campaign performance in real-time for more agile ad spend and creative adjustments.

Technical Outcomes

Data wrangling, grouping, and visualization of complex multi-channel marketing dataset

ROI calculation and actionable segmentation analysis

Interactive dashboards suitable for business review, stakeholder meetings.
